```python
res={}
```

```python
def get_history(tick):
    ticker = tick
    url_1 = f'https://financialmodelingprep.com/api/v3/historical-market-capitalization/{ticker}?limit=10000&from=2000-05-31&to=2005-05-31&apikey=FFGyH0tJPKANTOggUp6oDXjvLVOJdRoM'
    url_2 = f'https://financialmodelingprep.com/api/v3/historical-market-capitalization/{ticker}?limit=10000&from=2005-06-01&to=2010-05-31&apikey=FFGyH0tJPKANTOggUp6oDXjvLVOJdRoM'
    url_3 = f'https://financialmodelingprep.com/api/v3/historical-market-capitalization/{ticker}?limit=10000&from=2010-06-01&to=2015-05-31&apikey=FFGyH0tJPKANTOggUp6oDXjvLVOJdRoM'
    url_4 = f'https://financialmodelingprep.com/api/v3/historical-market-capitalization/{ticker}?limit=10000&from=2015-06-01&to=2020-05-31&apikey=FFGyH0tJPKANTOggUp6oDXjvLVOJdRoM'
    url_5 = f'https://financialmodelingprep.com/api/v3/historical-market-capitalization/{ticker}?limit=10000&from=2020-06-01&to=2024-05-31&apikey=FFGyH0tJPKANTOggUp6oDXjvLVOJdRoM'
    response1 = requests.get(url_1)
    response2 = requests.get(url_2)
    response3 = requests.get(url_3)
    response4 = requests.get(url_4)
    response5 = requests.get(url_5)
    data1 = response1.json()
    data2 = response2.json()
    data3 = response3.json()
    data4 = response4.json()
    data5 = response5.json()
    datatable1 = pd.DataFrame(data1).iloc[::-1]
    datatable2 = pd.DataFrame(data2).iloc[::-1]
    datatable3 = pd.DataFrame(data3).iloc[::-1]
    datatable4 = pd.DataFrame(data4).iloc[::-1]
    datatable5 = pd.DataFrame(data5).iloc[::-1]
    datatablefinal = pd.concat([datatable1, datatable2, datatable3, datatable4, datatable5],ignore_index=True)
    return datatablefinal
def run_history():
    for ticker in tickers:
        data = get_history(ticker)
        res[ticker]=data
    return res
res = run_history()
```

```python
monthly_df = monthly_dji_prices
returns_df = pd.DataFrame(index=monthly_df.index)
for ticker in monthly_df.columns:
    price_12_months_ago = monthly_df[ticker].shift(12)

# Calculate the price 1 month ago
    price_1_month_ago = monthly_df[ticker].shift(1)

# Calculate the momentum (R t-11 t-1) as the return from 11 months ago to 1 month ago
    returns_df[f'{ticker}_momentum'] = ((price_1_month_ago - price_12_months_ago) / price_12_months_ago) * 100
returns_various_times["mom"] = returns_df
pd.set_option('display.max_rows', None)
returns_various_times["mom"]
```
